# Data Cleaning and Preprocessing

```python
import pandas as pd
import ast
from datetime import datetime

# Convert datetime columns
user_df['yelp_since'] = pd.to_datetime(user_df['yelping_since'])
restaurant_review__df['date'] = pd.to_datetime(restaurant_review__df['date'])

# Fill missing values in categorical columns
user_df['elite'].fillna('Not Elite', inplace=True)

# Extract structured info from attributes column
restaurant_review__df['attributes'] = restaurant_review__df['attributes'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else {}
)

# Fill missing numerical values with zero
user_df.fillna(0, inplace=True)
restaurant_review__df.fillna(0, inplace=True)

# Drop duplicate reviews
restaurant_review__df.drop_duplicates(subset=['review_id'], inplace=True)

# Display cleaned dataset
print("User Data after Cleaning:")
display(user_df.head())

print("Restaurant Reviews Data after Cleaning:")
display(restaurant_review__df.head())
```

# Sentiment Analysis (NLP)

## Basic Sentiment Analysis (VADER)

```python
!pip install vaderSentiment # Installs the vaderSentiment package for sentiment analysis.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to compute sentiment score
def get_sentiment(text):
    if isinstance(text, str):
        return analyzer.polarity_scores(text)['compound']
    return 0

# Apply sentiment analysis
restaurant_review__df['sentiment_score'] = restaurant_review__df['text'].apply(get_sentiment)

# Categorize sentiments
restaurant_review__df['sentiment_category'] = restaurant_review__df['sentiment_score'].apply(
    lambda x: 'positive' if x > 0.1 else ('negative' if x < -0.1 else 'neutral')
)

# Display results
print("Sentiment Analysis Applied:")
display(restaurant_review__df[['text', 'sentiment_score', 'sentiment_category']].head())
```

## Advanced Sentiment Analysis (BERT)

```python
from transformers import pipeline, AutoTokenizer

# Load pre-trained sentiment analysis model and tokenizer
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
sentiment_pipeline = pipeline("sentiment-analysis", model=model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Function to apply sentiment analysis with truncation
def get_bert_sentiment(text):
    if isinstance(text, str):
        # Tokenize the text and get the model's prediction
        result = sentiment_pipeline(text, truncation=True, max_length=512)[0]
        return result['label']
    return 'neutral'

# Apply BERT-based sentiment analysis (first 1000 rows for efficiency)
restaurant_review__df['bert_sentiment'] = restaurant_review__df['text'][:1000].apply(get_bert_sentiment)

# Display results
display(restaurant_review__df[['text', 'bert_sentiment']].head())
```

# Building the Hybrid Recommendation System

```python
!pip install surprise
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split

# Load data for collaborative filtering
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(restaurant_review__df[['user_id', 'business_id', 'stars_y']], reader)

# Train-Test Split
trainset, testset = train_test_split(data, test_size=0.2)
algo = SVD()
algo.fit(trainset)

# Predict for a given user
user_id = 'nJIXm6xE1GBRap8qKFTLPw'
business_id = 'txyXRytGjwOXvS8s4sc-WA'
pred = algo.predict(user_id, business_id)
print(f"Predicted rating for User {user_id} on Business {business_id}: {pred.est}")
```

# User-User Collaborative Filtering

```python
from surprise import KNNBasic
from surprise import Dataset, Reader

# Load dataset
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(restaurant_review__df[['user_id', 'business_id', 'stars_y']], reader)

# Define similarity options
sim_options = {'name': 'cosine', 'user_based': True}

# Train user-user similarity model
algo = KNNBasic(sim_options=sim_options)
trainset = data.build_full_trainset()
algo.fit(trainset)

# Recommend for a specific user
uid = 'nJIXm6xE1GBRap8qKFTLPw'
iid = 'txyXRytGjwOXvS8s4sc-WA'
pred = algo.predict(uid, iid)
print(f"Predicted rating for User {uid} on Business {iid}: {pred.est}")
```

# Restaurant-Restaurant Content-Based Filtering

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Convert categories into a text-based feature
restaurant_review__df['categories_text'] = restaurant_review__df['categories'].apply(lambda x: ' '.join(x.split(',')))

# Vectorize text data
vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(restaurant_review__df['categories_text'])

# Compute cosine similarity between restaurants
restaurant_similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to recommend similar restaurants
def get_similar_restaurants(business_id, n=5):
    idx = restaurant_review__df.index[restaurant_review__df['business_id'] == business_id][0]
    similar_indices = restaurant_similarity[idx].argsort()[-(n+1):-1][::-1]
    return restaurant_review__df.iloc[similar_indices][['name', 'categories', 'stars_x']]

# Example: Find restaurants similar to Smoothie King
print(get_similar_restaurants('txyXRytGjwOXvS8s4sc-WA'))
```

# Hybrid Recommendation System

```python
def hybrid_recommend(user_id, business_id, n=5):
    user_based_rec = algo.predict(user_id, business_id).est  # Collaborative Filtering Score
    content_based_rec = get_similar_restaurants(business_id, n)  # Content-Based Score

    print(f"User-Based Predicted Rating: {user_based_rec}")
    print("Content-Based Recommended Restaurants:")
    print(content_based_rec)

# Example: Hybrid Recommendation for a user
hybrid_recommend('nJIXm6xE1GBRap8qKFTLPw', 'txyXRytGjwOXvS8s4sc-WA')
```
